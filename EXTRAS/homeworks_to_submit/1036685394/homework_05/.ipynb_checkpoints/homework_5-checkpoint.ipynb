```python
import json
import requests
```

```python
file= open('data.json')
data = json.load(file)
Country = "Spain"
URL = "https://inspirehep.net/api/institutions?q={}".format(Country)
```

```python
institution_link = requests.get(URL)
institution_link.json().get('hits').get('hits')
```

```python
for institution in institution_link.json().get('hits').get('hits'):
    i = institution.get('metadata').get('number_of_papers')
    if i>0:
        print(i)
        break
Legacy_ICN = institution.get('metadata').get('legacy_ICN')
```
